In electronics, a key domain of activity consists in building **sensors**. Inspired by the human senses (namely sight, smell, touch, taste, and hearing), sensors have a vast field of application. What fascinates about electronic sensors is their ability to overcome the domain where our human senses are sensitive. For example, in these domains, sensors can extend our sensing limit and hence, modify our perception:

- The human eye cannot perceive thermal (infra-red) radiations.
- Our ear can only hear sounds ranging typically from 20 to 20.000 hertz.
- Carbon monoxide has no smell.
- The Earth is in a continuous seismic activity.
- Water can be infected by bacteria or viruses.
- etc. 

During sensation, sense organs collect various stimuli (such as a sound or smell) for **transduction**, meaning transformation into a form that can be understood by the brain. The development of the sensor industry is based on translating, in electronic devices and signals, this sentence. The translation goes as follow: 

**Sensors** react to a physical phenomenon (electromagnetic wave, pressure, temperature, etc.) and convert it into a electric signal (current - voltage) that can be processed (by a computer for example) and interpreted by humans. Sensors are all around us - from smoke detectors to a simple ring bell. The integration of a high density of sensors into a single device is one of the challenge whose success translated into the smartphone (microphone, light sensor, camera, touch screen, orientation sensors, etc.).

## Step one: the circuit.

Most of the symbols composing this circuit are known to you. Most but not all of them: the central triangle makes its first appearance in the course. It is known as a **comparator**. It is one extremely simple compound regarding its concept, but a little more complicated regarding its actual implementation. Let us then focus on the concept. Here is the symbol for the comparator:

<img title="a title" alt="Alt text" src="../Figures/Cours 8/comparator.png" width="200">

A comparator, as its name suggests, compare two voltages ($U^{+}$ and $U^{-}$) between two of its input terminal and outputs a signal $S$. As a function of $\Delta U=U^{+}-U^{-}$, the comparator output is given as "1" (i.e. 5V) if $\Delta U$ is **positive** or "0" (i.e. 0V) if $\Delta U$ is **negative**. This peculiar response can be summed-up as follows:

<img title="a title" alt="Alt text" src="../Figures/Cours 8/comparator_function.png" width="400">

This comparator does not implement a function *per se*. Indeed any positive $\Delta U$ leads to the same value of $S$, namely "1" (there are no bijection... an infinite number of different $\Delta U$ correspond to the same $S$ value).This component is the bridge between the **analogic world** (continuous values) and the **digital** one (discrete values - "0" or "1"). It mimics an elementary **decision process**, i.e. Question: "Is $U^{+}$ bigger than $U^{-}$?". Answer: "Yes ("1") or No ("0")". Such decision processes, in more elaborate forms are present all across our society. For example, whenever there are two candidates for the same job. They continuously differ regarding many aspects. But a decision has to be taken and only one of them will get out of the decision process (which is here a black box, in comparison to the **deterministic** electronics comparator).

**Importantly**, the comparator needs to be powered appropriately to output e.g. 0 or 5V regardless of the amplitude of $\Delta U$ (ex cathedra comment).

Knowing the functional behavior of this comparator, let us understand the above circuit from the origin of time (when a battery is connected and imposes a voltage $U$ - in the bottom right part of the circuit)(not THE origin of time...). 

The instant the battery is connected, the value of the potential all across the circuit evolves. Considering that all the nodes are initially at zero volt, the circuit resumes as follow (detailed ex cathedra):

<img title="a title" alt="Alt text" src="../Figures/Cours 8/state_1.png" width="200">

Solving for $U^{+}$^(left to you as an exercise), one can find that $U^{+}=\frac{1}{3}U$, quasi instantaneously after the battery is plugged. This means that $U^{+}$ is **bigger than** $V$ (one has that $V=U^{-}$ but $V$ is used as a symbol to stay consistent with course 7 - $V$ the voltage at the terminals of the capacitor). As a consequence, the comparator **switches** to the "1" state. The voltage it reaches is the maximum voltage acrosse the circuit (that powers it) ,i.e. $U$. 

This has two **consequences**:

- First, the capacitor **charges** with the well known "charging pattern" (as presented in course 7) with a time constant $\tau$ given by the product $RC$. 
- Second, the voltage $U^{+}$ changes. Indeed, the "new" equivalent circuit is the following:

<img title="a title" alt="Alt text" src="../Figures/Cours 8/state_2.png" width="200">

Solving for $U^{+}$^(left to you as an exercise), one can find that $U^{+}=\frac{2}{3}U$. This new value of $U^{+}$ means that the comparator will not switch unless $V$ reaches a value bigger that $\frac{2}{3}U$! So the capacitor charges up to this value for $V$. Once $V$ becomes bigger (really slightly - see comment below) than $\frac{2}{3}U$, the comparator **switches** for $U$ (the "1" state) to 0. This has again two consequences:

- First the capacitor **discharges** and the stored energy is dissipated in $R$.
- Second, $U^{+}$ changes back to $\frac{1}{3}U$.

The interesting feature of this circuit is now a bit clearer. Indeed, as the capacitor discharges, $V$ becomes lower than $\frac{1}{3}U$. Just after crossing this critical value, the comparator switches back to the value $U$ since $V<U^{+}$. 

The loop goes on forever. In the **steady state**, the voltage $V(t)$ goes as follow:

<img title="a title" alt="Alt text" src="../Figures/Cours 8/capa_voltage.jpg" width="600">

An important question you might wonder is: how fast does the comparator **switches** between the "0" and the "1" state?

By looking back to the comparator "function" (see above), one could think that the slope (connecting the "0" - blue curve - to the "1" -red curve) is **infinite**. Actually, modern-day comparators present (insanely) high slope. The value of this slope is typically on the order of $10^{5}$

$$\frac{\Delta S}{\Delta U}\sim 10^{5}$$

Regarding the amount of **time**, one has to consider the **slew rate**, typically on the order of $300\; mV/\mu s$. To ramp up from 0V to 5V, the comparator takes **only** a few micro seconds ($\sim5/0.3\;\sim15\;\mu s$).

**The signal's frequency $f$ is proportional to the rate at which the capacitor charges - hence, proportional to $\frac{1}{RC}$**.

Here is the first equation to understand today:

$$f\propto\frac{1}{RC}$$

# Step 2: the sensor

The aforementioned circuit allows us to consider a **signal** (rising/falling) whose frequency vary as the inverse of the product $RC$. Suppose that a physical phenomenon that one wishes to caracterise influences the value of either the resistance (leading to a resistive sensor) or the capacitance of the capacitor (leading to a capacitive sensor). For sensing application, there must exist a **one to one** correspondancy (bijection) between the value of the physical phenomenon (that one plans to characterize) and the value of the component that is considered as a sensor. To go further with the idea of **measuring a water level**, suppose the following device:

<img title="a title" alt="Alt text" src="../Figures/Cours 8/water_box.png" width="300">

A box is sandwiched between the two parallel plates (whose total area is $A=L\times H$, where $H$ is the length of the edge perpendicular to the water level and $L$, the length of the other edge of the plate) of a capacitor. When the box is **empty**, the **dielectric** present between the plates is air (relative permittivity - $\epsilon_{air}=1$). When the box is **full of water**, the **dielectric** present between the plates is... water (relative permittivity - $\epsilon_{water}=80$). On the one hand (box empty), we have:

$$C=1\frac{A}{d}$$

where $d$ is the distance between the plates. And on the other (box full), we have:

$$C=80\frac{A}{d}$$

What is the value in between? (try to do it as an exercise)

In the situation between full and empty, one have two capacitor in **parallel**. It can be represented as follows:

<img title="a title" alt="Alt text" src="../Figures/Cours 8/model.png" width="600">

The equivalent capacitor (of parallel capacitors) is simply the sum of the two.

$$C_{tot}=C_{air}+C_{water}=\epsilon_{air}\frac{(H-h)L}{d}+\epsilon_{water}\frac{hL}{d}=\frac{L}{d}\left(\epsilon_{air}H+(\epsilon_{water}-\epsilon_{air})h\right)$$

Depending on the water level, the capacitance vary and the bijection is this simple linear (affine) relation. The slope of this relation is given by:

$$\frac{L}{d}(\epsilon_{water}-\epsilon_{air})$$

Hopefully, there is a large difference between water and air, regarding the relative permittivity $\epsilon$.

Now, depending on the capacitance value, one have a specific frequency thanks to the above circuit. Before going on the market, such sensor would need to be **calibrated** (Important ex cathedra explaination).

In general, **capacitive sensor** are not only based on a variation of relative permittivity (this was only one example). But... what other variation can be sensed?

To understand that, let us use infinitesimal computations: namely, the partial derivative **chain rule**.

Given that:

$$C=\epsilon\frac{A}{d}$$

a finite $C$ variation (written as $\Delta C$) is given by:

$$\Delta C=\frac{\partial C}{\partial\epsilon}\Delta\epsilon+\frac{\partial C}{\partial A}\Delta A+\frac{\partial C}{\partial d}\Delta d$$

with 

$$\frac{\partial C}{\partial\epsilon}=\frac{A}{d}$$

$$\frac{\partial C}{\partial A}=\frac{\epsilon}{d}$$

and

$$\frac{\partial C}{\partial d}=-\frac{\epsilon A}{d^{2}}$$

Now, suppose that we know a reference value $C_{0}=\frac{\epsilon_{0}A_{0}}{d_{0}}$.

The **absolute** variation **from** this known value is given by:

$$\Delta C=\frac{A_{0}}{d_{0}}\Delta\epsilon+\frac{\epsilon_{0}}{d_{0}}\Delta A-\frac{\epsilon_{0} A_{0}}{d_{0}^{2}}\Delta d$$

From which we can find the **relative variation**:

$$\frac{\Delta C}{C_{0}}=\frac{\Delta \epsilon}{\epsilon_{0}}+\frac{\Delta A}{A_{0}}-\frac{\Delta d}{d_{0}}$$

From this equation, one can **imagine** all sort of sensors!
