# Back to basics

Before digging in the amazing (yet possibly frightening) world of electronics, let us discuss a simple concept: the **circuit**. Phenomenologically, a circuit presents a **starting** point **connected** to an **ending** point. On the **path** connecting both points, there is an **object** in a particular **motion**. Let us consider, for example, the motion of a cyclist on a bike. Suppose you started from home and came by bike to the course. You start by picturing the path in your head. Nevertheless, the path that you will take is not **exactly** the one that you considered. On the **actual** path, you will maybe:

- encountered unexpected closed street and be forced to make a by-pass. 
- have to slow down and slalom between people in a crowded street.
- adapt your speed to the road nature (grass, concrete, etc.) 

Due to all these speed variation, you will gain or dissipate **energy**. To climb a hill, you will have to convert kinetic energy to potential energy. If you encounter a stop sign, by braking, you will turn kinetic energy into heat. Etc.

Let us elaborate on the comprehension of circuits by looking to a **water** circuit.

There is something we did not mentioned yet, the water **pipes**: what will they be analogous to ? Electric **wires** realized from electrically **conductive** material (such as copper, gold, silver, etc.).

# Unit talk

# Electronics: from keywords to physics

The field of **electronics** is a branch of physics and electrical engineering aiming at manipulating **electrons** to realize specific **tasks** by using **electronic devices**. Electronics is primarily concerned with phenomena other than simple **conduction**, such as emission of electrons (e.g. in cathode-ray tube), storage of electrical charge (e.g. batteries), light-matter interactions (e.g. solar panels), the effects of electrical fields on the conduction of electrons through a circuit, amplification, rectification and manipulation of electric signals such as voltage or current by **design of circuits**. Questions such as **transferring energy** or **transmitting a signal** are at the root of electronics concerns.

Electronics also encompasses the application of **fundamental principles** (translated into **Maxwell's equations**) to the **construction of devices**, known as electronic devices. 

Electronics can be studied with different prospectives (to name a few):

- the **motions of electrons** (e.g. acceleration, oscillation etc.) in an electronic device (more specifically, the influence of this motion on voltage, current, electric and magnetic fields)
- the **materials** used to construct the device
- the **design** of circuits (integrated circuits, printed board circuits)
- the value chain of electronic products

Electronics has hugely influenced the development of modern society (Can you name a few examples?). Before getting too deep into the connections between electronics and the other knowledge fields where it plays a crucial role (**informatics**, **communication** technologies), let us start by listing the The **roadmap** guiding the development of modern electronic devices. 

They need to:

- work at **high frequency**
- be **low power**
- be **affordable**
- be **sustainable**

### Printed board circuit (PCB)

A **printed** circuit board is a medium used in electrical and electronic engineering to connect electronic components to one another in a controlled manner. It takes the form of a **laminated** sandwich structure of conductive and insulating layers: each of the conductive layers is designed with an artwork pattern of traces, planes and other features (**similar to wires on a flat surface**) etched from one or more sheet layers of **copper** laminated onto and/or between sheet layers of a non-conductive substrate. Electrical components may be fixed to conductive pads on the outer layers in the shape designed to accept the component's terminals, generally by means of soldering, to both electrically connect and mechanically fasten them to it.

#### The integrated circuit

An integrated circuit or monolithic integrated circuit (also referred to as an **IC**, a chip, or a microchip) is a set of electronic circuits on one small flat piece (or "chip") of **semiconductor material**, usually silicon. Large numbers of **miniaturized transistors** and other electronic components are integrated together on the chip. This results in circuits that are orders of magnitude smaller, faster, and less expensive than those constructed of **discrete components**, allowing a large transistor count. The IC's **mass production capability**, reliability, and building-block approach to integrated circuit design has ensured the rapid adoption of standardized ICs in place of designs using discrete transistors. ICs are now used in virtually all electronic equipment and have revolutionized the world of electronics.

<div class="alert alert-block alert-info">
In microelectronics, a dual in-line package (DIP or DIL), is an electronic component <b>package</b> with a rectangular housing and two parallel rows of electrical connecting pins. The package may be through-hole mounted to a printed circuit board (PCB) or inserted in a socket.
</div>

It has eight **pins**. The pins on an IC chip provide connections to the tiny integrated circuits inside of your electronics. The pins are part of the **lead frame** which is connected to the **die** thanks to **bonding wires**. A lead frame (pronounced /lid/) is a metal structure inside a chip package that carries signals from the die to the outside world.

The **mold resin** protects the **die** from humidity, moisture, scratches etc. This mold resin (epoxy polymer) is extremely robust and prevent from accessing easily to the die itself. Nevertheless, sophisticated techniques (chemical etching, laser cutting etc.) allows **hackers** to remove the epoxy packaging from chips to expose the die and reveal (with an optical microscope) its **architecture**. Such process is called **decapping ICs**. This (illegal) process is the ultimate hardware technique to **hack** someone's phone. For the curious, let us mention the juridic battle between the company Apple and the FBI. In an investigation, one technique investigators could theoretically employ “decapping,” by physically taking out the phone’s memory chip and use a combination of acid and a tiny laser drill to extract bits of data (https://en.wikipedia.org/wiki/FBI–Apple_encryption_dispute).

The **die** is a small block of semiconducting material on which a given functional circuit is fabricated. Typically, integrated circuits are produced in large batches on a single wafer of electronic-grade **silicon** or other semiconductor (such as Gallium Arsenide) through processes such as **photo-lithography**. The **wafer** is cut (diced) into many pieces, each containing one copy of the circuit. Each of these pieces is called a die.

A silicon **wafer** (disks visible on the right) is a slice of an **ingot** of silicon (on the left in the following picture). The slice is polished to get "flat" surface (extremely small roughness). The standard diameters are the polished 12 or 6 inches silicon wafers.
