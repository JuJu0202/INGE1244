This exercise session is dedicated to the fact that two wires in front of each other also have a capacitance. Indeed, eventhough these are not parallel flat surface, if one applies a difference of potential $V$ between the wires, there will be a finite charge $+Q$ (on one of the wires, say the red one) and a finite opposite charge $-Q$ on the other one (say the blue one).

The capacitance can then be expressed as:

$$C=\frac{Q}{V}$$

Another type of cable is the **coaxial cable**. Such type of cable is ubiquitous in electronics and especially **telecommunications**. Presenting the actual properties and use case of the coaxial cable is far beyond the scope of this exercise session but let us factual present such cable.

<img title="a title" alt="Alt text" src="../Figures/Exercise 6/coaxial.png" width="500">

It consists of on inner copper core (stranded or solid) surrounded by an insulator. The insulator is then encircled by a copper cylindrical shell ("meshed" copper on the above figure) which is covered with an outside insulator. From a electric point of view, a coaxial cable can be seen as a **capacitor**. Indeed it corresponds to the definition of the capacitor: **two conductive surfaces in front of one another**. If the inner wire is charged with an electric charge $+Q$, the outer shell gets charges, by electrostatic induction, with a charge $-Q$. The **spacing** between these **charges** (given by the thickness of the insulator - that is a dielectric) leads to a system that presents **potential energy** $V$. The capacitance of the capacitor is defined by the ratio between $Q$ and $V$ (as usual). For the following system,

<img title="a title" alt="Alt text" src="../Figures/Exercise 6/cylinder_capa.jpg" width="500">

the capacitance is given by:

$$C=\frac{2\pi\epsilon}{ln\left(\frac{b}{a}\right)}l$$

where $\epsilon$ is the permittivity of the inner insulator ($\epsilon=\epsilon_{R}\epsilon_{0}$), $b$ the outer cylinder radius, $a$ the inner cylinder radius, $ln$ is the natural logarithm and $l$ the length of the considered cable. If you want some more information about that system, have a look to : https://www.youtube.com/watch?v=4VSMi5oxz_U

For fixed parameters (permittivity and radius), the law is extremely simple as the capacitance is directly proportional to the **length**:

$$C\propto l$$

# An elementary sensor: data acquisition and analysis

During last course, we discussed an elementary circuit that can be used to collect information from an elementary **transducer**, i.e. a capacitor whose capacitance variations can be associated to a physical phenomenon (e.g. the water level in a box).

This process is a fundamental step in electronics, named, **data acquisition**.

Data acquisition is the process of **sampling** signals that **measure** real-world physical conditions and converting the resulting samples into digital numeric values that can be manipulated by a computer. 

Data acquisition systems typically **convert** analog waveforms into digital values for processing. This last sentence is the purpose of this exercise session. 

The components of data acquisition systems include:

- **A sensor**, to convert a physical phenomenon to electrical signals.
- **Signal conditioning circuitry**, to convert sensor signals into a form that can be converted to digital values.
- Analog-to-digital converters, to convert conditioned sensor signals to digital values.

<img title="a title" alt="Alt text" src="../Figures/Exercise 6/data_acqu.jpg" width="600">

Data acquisition applications are usually controlled by software programs developed using various general purpose programming languages such as Assembly, BASIC, C, C++, C#, Fortran, Java, LabVIEW, Pascal, Python etc. These are widely used in the industry and you will most probably encounter one of them in your career.
