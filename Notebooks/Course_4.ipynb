Components can be classified as **passive**, **active**, or **electromechanic**. Let us define the terms as used in circuit analysis as:

- **Active components** rely on a source of energy (usually from the DC circuit, which we have chosen to ignore) and usually can inject power into a circuit. Active components include amplifying components such as transistors, triode vacuum tubes (valves), and tunnel diodes.
- **Passive components** cannot introduce net energy into the circuit. They also cannot rely on a source of power, except for what is available from the circuit they are connected to. As a consequence, they cannot amplify (increase the power of a signal), although they may increase a voltage or current (such as is done by a transformer or resonant circuit). Passive components include two-terminal components such as resistors, capacitors, inductors, and transformers.
- **Electromechanical components** can carry out electrical operations by using moving parts or by using electrical connections

These components are used in two types of circuits: **direct current** (DC) circuits or **alternating current** (AC) circuits.

- **Direct current** is one-directional flow of electric charge  (an electrochemical cell - battery - is a prime example of DC power supply). The electric current flows in a constant direction, distinguishing it from alternating current.
- **Alternating current** is an electric current which **periodically** reverses direction and changes its magnitude continuously with time.

#  Ohm's law 

Note: Chapter 20.2 of your reference book complements these notes.

In an ordinary metal such as copper or aluminum, **some of the electrons** are free to move within the conductive material (they are called **conductive electrons**). The charges move randomly within the material in homogeneously distributed direction. Their velocity in this motion is constant and is called the **Fermi** velocity. The signification of this velocity is buried into quantum mechanics and the definition of the **Fermi energy**. For copper, the Fermi **energy** ($E_{F}$) of electrons is about 7.00 **eV** which gives a fermi velocity ($v_{F}$ - given by the relation $E_{F}=\frac{m_{e}v_{F}^{2}}{2}$) of about $1.57\times10^{6}\;m/s$. Under the **free electron model**, the electrons in a metal can be considered in the same way one could picture the atoms or molecules in a **perfect** gas. The number density of conduction electrons in metals ranges between approximately $10^{28}$ and $10^{29}$ electrons/$m^{3}$, which is also the typical density of atoms in ordinary solid matter. Even if they are called *free*, the electrons do not escape from the conductive material because they are attracted to the positive ions of the material (a material, if not charged... is electricaly neutral).

Electrons in the material can be pictured as **ping-pong balls** experiencing many **collisions**. These collisions make electrons **randomly** change their direction. Between two collisions with the much **heavier**, nearly **stationary** ions of the material, electrons move in a straight line at a constant velocity $v_{F}$. 

Here is how it can be pictured:

<img title="a title" alt="Alt text" src="../Figures/Cours 4/random_path.png" width="200">

The distance between point a and point b is not exactly zero for a given trajectory but **averaged** over all the electron trajectories, this distance is indentically **zero**: There are **no** charge displacement within the material. 

Now, consider that a constant electric field $\vec{E}$ is established inside the conductor. How to establish this field is a question adressed later. As presented in INGE1243, an electric field causes a force $\vec{F}=q\vec{E}$ (where $q$ is the electric charge of the particle). Under a constant force, in **vacuum**, the charge would undergo a steady acceleration and its velocity would increase (more and more). The change of velocity would be given by the **work-energy theorem**: the force applied on a given distance corresponds to a work which changes the kinetic energy of the particle. The electric field does **work** on the charge. The resulting kinetic energy is **transfered** to the ions in the material during the **inelastic** collision. The ions then vibrate more about their equilibrium positions in the crystalline structure of the conductor. This energy transfer increases the average vibrational energy of the ions and therefore, the **temperature of the material**. The net effect of the electric field is that there also is a **very slow** motion of the electrons called the **drift velocity** ($v_{d}$). The displacement has now to be measured between point a and point c of the following figure (which is **no more zero**):

<img title="a title" alt="Alt text" src="../Figures/Cours 4/random_path_E.png" width="400">

This **drift velocity** causes net displacement of electric charge, namely, **an electric current** ("I" expressed in ampere). We can obtain an expression for the relationship between current and drift velocity by considering the number of free charges in a segment of wire, as illustrated in the following figure:

<img title="a title" alt="Alt text" src="../Figures/Cours 4/drift_velocity.jpg" width="500">

The number of free charges per unit volume is represented by the symbol $n$ and depends on the material. The shaded segment has a volume $Ax$, so that the number of free charges in it is $nAx$. The charge $\Delta Q$ in this segment is thus $qnAx$, where $q$ is the amount of charge on each **charge carrier** (here, electrons, where $q=-1.6\times10^{-19}\;C$)

Current is "charge moved per unit time"; thus, if all the original charges move out of this segment in time $\Delta t$, the current is:

$$I=\frac{\Delta Q}{\Delta t}=\frac{qnAx}{\Delta t}$$

where $\frac{x}{\Delta t}$ defines the **drift velocity**. Rearranging the terms leads to:

$$I=nqAv_{d}$$

where $I$ is the current through a wire of cross-sectional area $A$ made of a material with a free charge density $n$. The carriers of the current each have charge $q$ and move with a drift velocity of magnitude $v_{d}$. Note that, as electrons carry a negative charge, $v_{d}$ is opposite to $\vec{E}$.

Front this current definition, we can also define the **current density** (noted $J$) as:

$$J=\frac{I}{A}\;A/m^{2}$$

### Resistivity 

Now comes the interesting question: how can we double (say) the current density in a given material? Intuitivelly, you would argue that one need to increase the electric field (causing the drift)... and you would be right! But  The answer is attributed to **Georg Simon Ohm** who demonstrated that the ratio between the electric field (cause) and the current density (consequence) was **constant** for a **given** material and was called the **resistivity** (noted $\rho$). The greater the resistivity ($\rho$), the greater the field ($E$) to cause a given current density (J). We have:

$$E=\rho J$$

This quantity allows to distinguish between **metals** and **insulators** (as the resistivities of insulators are greater than those of metals by an enormous factor of $10^{22}$!). The units of the resistivity are $(volt/meter)/(ampere/meter\;square)$ = $\frac{Vm}{A}$. As we will see shortly, $volt/ampere$ is express in **ohms** (noted with the letter $\Omega$). Thus, the resistivity expresses in $\Omega m$. The reciprocal of resistivity is **conductivity**.

### Resistance

To cause an electric field into a material, a potential **difference** $\Delta V$ has to exist between high and a low voltage locations. Let us consider (with the figure presented below) that the first face of this cylindrical conductor is at the high potential and that the second face is at the low potential - causing the electric field $E$. 

<img title="a title" alt="Alt text" src="../Figures/Cours 4/cylinder.png" width="400">

As we know from INGE1243, 

$$E=\frac{\Delta V}{L}$$

The material has a given resistivity $\rho$ leading to a current density $J$ given by:

$$J=\frac{I}{A}=\frac{E}{\rho}=\frac{\Delta V}{\rho L}$$

which gives:

$$\Delta V=\frac{\rho L}{A}I$$

The ratio between the potential diffence and the current is called the **resistance** (noted with the letter $R$):

$$R=\frac{\rho L}{A}$$

### Power relation

$$P=\Delta V I$$
