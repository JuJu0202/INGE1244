# Potential energy : a remark

A central concept in this course about electronics (and the previous course INGE1243 as well) is the one of **potential energy**. This concept has been investigated in many physical systems up to now. Let us review those that we encountered. 

### The "ghost" tennis ball

Suppose that a ghost tennis ball is standy at the center of the Earth. The "ghost" tennis ball keeps every of its physical properties (its mass ($m_{ball}$), its drag coefficient, its fluffiness, etc.) **but** it can move through any form of matter without encountering any reaction force. Strictly **at** the center, the ball has **no** potential energy (important to note). This means that, to move the ball from the center of the Earth to the surface of the Earth, work needs to be done. Indeed, the ball feels the **gravitational force**... but so does the **Earth** (Newton's **action-reaction** third principle). Suppose that such displacement of the ball is realized (by some sort of almighty god that provides the energy). Doing so, if the ball is displaced **at constant speed** (sum of the force been zero), the "Earth-ball" system as a **whole** (a two body system) gains **only** potential energy (P.E.) whose value is given by:

$$P.E.=Work=Force\times distance$$

In such a gravitational system, the force is given by the universal law of attraction (Newton) and its amplitude **depends** on the distance from the center. This work is really an **integral** given as follows:

$$Work=\int_{0}^{R}\vec{F}(r)\cdot\vec{dr}=\int_{0}^{R}F(r)dr$$

where $R$ stands for the radius of the Earth, and $F$ is the gravitational force. The expression of the gravitational force inside a solid sphere is not a straightforward expression because the expression of the mass that needs to be taken into account in the expression of the gravitational force depends on the enclosed volume as well as on the density of the Earth... If you want to dig (literally) any further, I let you consult the exercise session 9 of INGE1243 ("Journey to the center of the Earth"). For the rest of this notebook, let us suppose that the work is known and its value is $W$ (joules) - the **only** energy available to this two body system.

<div class="alert alert-block alert-info">
For your information <b>only</b>, we can estimate this potential with a "simpler" method. Let us find another reference point where the ball has a reference potential energy of **zero** (the same value as when it stands at the center). Where would this point be? At an **infinite** distance from the center of the Earth. The trick is now to note that the work that needs to be **done** to bring the ball from the center of the Earth to the Earth surface is opposite in sign but **equal** in magnitude to the work corresponding to the fact of bringing the ball, **at constant speed**, from infinity to the Earth surface. We have:

$$P.E.=Work=\int_{+\infty}^{R}F(r)dr$$

where this time, $F(r)$ is simply given by:

$$F(r)=G\frac{m_{ball}M_{Earth}}{r^{2}}$$

$$Work=\int_{\infty}^{R}G\frac{m_{ball}M_{Earth}}{r^{2}}dr=-\left.{G\frac{m_{ball}M_{Earth}}{r}}\right|_{\infty }^{R}=-{G\frac{m_{ball}M_{Earth}}{R}}$$

In the common situation where a much smaller mass ($m_{ball}$) is moving near the surface of a much larger object (the Earth) the gravitational field is nearly constant and so the expression for gravitational energy can be considerably simplified. The change in potential energy $\Delta P.E.$ when moving from the surface (a distance $R$ from the center) to a height $h$ above the surface is given by:

$$\Delta P.E.=G\frac{m_{ball}M_{Earth}}{R}-G\frac{m_{ball}M_{Earth}}{R+h}=G\frac{m_{ball}M_{Earth}}{R}\left(1-\frac{1}{1+\frac{h}{R}}\right)$$

If $\frac{h}{R}$ is small, the expression $\frac{1}{1+\frac{h}{R}}$ is extremely close to $1-\frac{h}{R}$ (indeed, $1+\frac{h}{R}$ is extremely close to $e^{\frac{h}{R}}$ so, $\frac{1}{1+\frac{h}{R}}\approx \frac{1}{e^{\frac{h}{R}}}\approx e^{-\frac{h}{R}}\approx 1-\frac{h}{R}$. For example, if $\frac{h}{R}=0.001$, $e^{\frac{h}{R}}=1.0010005...\approx 1+0.001=1+\frac{h}{R}$. Recall that the radius of the Earth $R$ equals approximately $6.3\times 10^{6}m$). Hence, the approximation we are going to make in the following step is incredibly good:

$$\Delta P.E.=G\frac{m_{ball}M_{Earth}}{R}\left(1-\frac{1}{1+\frac{h}{R}}\right)\approx G\frac{m_{ball}M_{Earth}}{R}\left(1-(1-\frac{h}{R})\right)=m_{ball}\frac{GM_{Earth}}{R^{2}}h$$

Let us write $g=\frac{GM_{Earth}}{R^{2}}$, and, taking the average radius of the Earth, one find $g=9.8\;m/s^{2}$ and the potential to be simply given by:

$$\Delta P.E.=mgh$$
</div>

With this potential energy ($W$) available to the two body system, let us now apply the familiar **work energy theorem**.

$$\Delta K.E.=W$$

**But**! Both the Earth and the tennis ball will moves. So both objects gain kinetic energy. The actual way to apply this theorem is to state that, the change in K.E. of the Earth **plus** the change in K.E. of the ball **equals** te available energy $W$. As both objects start from **rest**, we have that the final velocities of the Earth (written $v_{1}$) and the ghost tennis ball (written $v_{2}$) satisfy:

$$\frac{M_{earth}v_{1}^{2}}{2}+\frac{m_{ball}v_{2}^{2}}{2}=W$$

In this equation, we notice **two unknowns** ($v_{1}$ and $v_{2}$). We cannot solve the problem without an additional equation.  
Hopefully, since no external force is applied to the system, the sum of the momenta (recall that $p=mv$) of the two bodies must be **zero** and both momenta have then to be equal in magnitude:

$$M_{earth}v_{1}=m_{ball}v_{2}$$

Substituting for $v_{1}$ into the work energy equation, one finds:

$$\frac{m_{ball}}{M_{Earth}}(\frac{m_{ball}v_{2}^{2}}{2})+\frac{m_{ball}v_{2}^{2}}{2}=\frac{m_{ball}v_{2}^{2}}{2}(1+\frac{m_{ball}}{M_{Earth}})=W$$

Due to the fact that the mass of the Earth is **huge**, one can neglect the first term of this equation and find the familiar:

$$\frac{m_{ball}v_{2}^{2}}{2}\approx W$$

The **main message** is the following: when lifting a tennis ball from the ground, even thought, it is the Earth-ball system as a **whole** that gains potential energy, the problem can be apprehended, to an extremelly good approximation as if the tennis ball gained all of $W$! We do **not** have to bother about the Earth in our description.

In the case when a tennis ball is positioned a distance $h$ from the surface of the Earth, if we launch the ball, the work energy theorem simply writes:

$$\Delta K.E.=P.E.=mgh$$

(since $\frac{h}{R}<<<1$ and $\frac{m_{ball}}{M_{Earth}}<<<1$)

Let us review the different source of potential energy that we encountered so far.

- For a tennis ball, the P.E. (in joules) is proportional to the distance (distance exponent **1**)
- For a spring attached to the ceiling, the P.E. (in joules) is proportional to the stretched distance squared (distance exponent **2**)
- For a system of two point charges, the potential energy (in joules per coulomb) of the two charges system (as a whole) is proportional to the inverse distance (distance exponent **-1**)
- For a system of two (infinite) metalic plates parallel to one another, charged respectivelly with $+Q$ and $-Q$, the P.E. (in joules per coulomb) is proportional to the distance ($d$) separating the two plate (distance exponent **1**).

Lifting a tennis ball and separating two parallel charged plates is really **conceptually** extremely close.

Note that when the two plates are touching one another, the P.E. is zero because the positive charge exactly compensate for the negative one.

# The RC-circuit

The capacitance of the capacitor is expressed in coulomb per volts (since $C=\frac{Q}{V}$). This unit is so ubiquitous in electronics that it was given its own unit: **the Farad** (named after the physicist Michael Faraday (1791-1867)).

As you might conceptualise, **no current** flows inside the capacitor. It does not dissipate energy, it just **stores** it. 

<div class="alert alert-block alert-info">
The energy stored in a capacitor equals the total **work** needed to bring a total charge $+Q$ on one plate (and induce $-Q$ to the other) to the potential $V$ (starting from the uncharged plates). The infinitesimal work (for an infinitesimal amount of charge $dq$) is given by:

$$dW=\frac{q}{C}dq$$

where $C$ is the capacitance. 

Eventually we have:

$$W=\int_{0}^{Q}dW=\int_{0}^{Q}\frac{q}{C}dq=\frac{Q^{2}}{2C}=\frac{1}{2}CV^{2}$$
</div>

Once charged (fixed amount of charge at the imposed potential), the capacitor acts as an **open circuit**. At this step of your reading, you might have questions about the current associated with such object. 

The current **associated to** a capacitor is due to the charge **time-variation** it induces during its charge or discharge.

Indeed, by differentiating with respect to time, 

$$Q=CV\leftrightarrow\frac{dQ}{dt}=C\frac{dV}{dt}$$

The term $\frac{dQ}{dt}$ is no other thing than a **current** $I$ (charge per unit of time). This gives us the differential law of the capacitor:

$$I=C\frac{dV}{dt}$$

Now that we elaborated on the **capacitor**, it is time to realize a **circuit** with this object. Let us consider the following circuit composed of: a battery (on the left), 2 switches (labelled A and B), a resistor (with a resistance of $R$ ohms) and a capacitor (with a capacitance $C$)(note also the symbol for the capacitor that symbolizes two plates). Three points (labelled 0, 1 and 2) are taken as reference points (nodes) where the potential is specified.

<img title="a title" alt="Alt text" src="../Figures/Cours 7/RC.png" width="400">

Let us first wonder what happens when switch A is **closed** (at time $t=0$ s). The potential difference due to the battery (noted $V_{0}$ - e.g. 12V) is imposed to the resistor and the capacitor connected in series. As a consequence, due to charge conservation, any current $I(t)$ (the current now depends on time, you will see why) flowing through the resistor, "flows" also through the capacitor. 

But, **on the one hand**, the current throught the resistor (Ohm's law) is given by:

$$I(t)=\frac{V_{1}-V_{2}(t)}{R}=\frac{V_{0}-V_{2}(t)}{R}$$

Note that, when the switch is closed (quasi-instantaneously), $V_{1}$ is raised to the constant value of $V_{0}$. $V_{2}$ depends on time. This time dependency is due to the fact that the energy cannot be stored instantaneously inside the capacitor (the battery cannot deliver the current needed to charge completely the capacitor instantaneously). 

**On the other hand**, the current throught the capacitor is given by:

$$I(t)=C\frac{dV_{2}(t)}{dt}$$

The initial condition is given by $V_{2}(t=0)=0$ and, for the long times (considered as infinite), the capacitor is charged and $V_{2}(t=+\infty)=V_{0}$. It is important that you understand well these two boundary conditions.

Equating both expressions for the current, we have the following first order differential equation (whose solution is uniquely specified by the boundary conditions):

$$V_{0}=V_{2}(t)+RC\frac{dV_{2}(t)}{dt}$$

The solution to such equation is the superposition (sum) of a particular solution (noted $V_{2}^{p}$) and an homogeneous solution (noted $V_{2}^{h}$)(cfr. your math courses).

On the one hand, the particular solution is given by $V_{2}^{p}(t)=V_{0}$ (indeed, check in the equation). On the other hand, the homogeneous solution is obtained by solving the *characteristic polynomial* associated to the equation. Depending on your notations, the characteristic polynomial is here $1+RC\lambda=0\leftrightarrow \lambda=\frac{-1}{RC}$. The homogeneous solution is given by:

$$V_{2}^{h}(t)=C\;exp\left(\frac{-t}{RC}\right)$$

where $C$ is an undetermined constant of integration and $exp()$ is the exponential function with base $e$ (=2.71...). To "fix" the solution, given by:

$$V_{2}(t)=V_{2}^{p}+V_{2}^{h}=V_{0}+C\;exp\left(\frac{-t}{RC}\right)$$

let us use the boundary condition stating that $V_{2}(t=0)=0$ (the other one being already fullfilled). This leads to:

$$V_{2}(t=0)=0=V_{0}+C\leftrightarrow C=-V_{0}$$

The final solution is then:

$$V_{2}(t)=V_{0}\left(1-\;exp\left(\frac{-t}{RC}\right)\right)$$

As we can see, thanks to the solution, the current immediately starts to flow within the resistor ($I(t=0^{+})=\frac{V_{0}}{R}$ - the notation $0^{+}$ signifies the instant a time infinitesimally close to 0 but positive) and an initial burst of current starts to charge the capacitor (rising progressively the potential $V_{2}$). Doing so, the current inside the resistor naturally decreases since the difference of potential at its terminals vanishes exponentially over time. 

Now that this is clear what difference of potential is worth discussing, let us drop the subscript index "2" and write $V_{2}(t)$ simply as $V(t)$. Also, the product $RC$ has the physical dimension of a **time**. It is called the **time constant** and (in electronic) is refered to with the greek letter tau ($\tau$). 

How does the potential difference $V(t)$ look with respect to time? The answer can be found on the following graph. 

<img title="a title" alt="Alt text" src="../Figures/Cours 7/RC_time.png" width="800">

The red curve represents $V(t)$. The blue curve (step function) represents the voltage that one tries to impose to the capacitor to charge it. The **counterintuitive point** about this physical phenomenon is that, **no matter** how high is the value of the potential $V_{0}$, it needs about three time constant for the capacitor to be (almost) fully charged (at $t=3\tau$, $V(t)/V_{0}=(1-exp(-3))=0.9502...$). Pause on this last sentence because it is truely counterintuitive for many students. It is only after an infinite time that $V(t)=V_{0}$. 

Now that the capacitor is charged, with a stored energy $\frac{CV_{0}^{2}}{2}$, let us re-open switch "A". What happens? **Nothing**. The capacitor is charged and no circuit is closed, so no current can flow: the energy is **stored**.

Now comes the time to close switch B. In such situation, the circuit becomes equivalent to:

<img title="a title" alt="Alt text" src="../Figures/Cours 7/RC_discharge.png" width="200">

In this scenario, the energy stored in the capacitor gets dissipated entirely through the resistor. The integral of the **power** ($p(t)=V(t)I(t)=RI^{2}(t)$) dissipated through the transistor is equal to the stored energy $\frac{CV^{2}(t)}{2}$, as follows (the switch is considered to be closed at a need reference time origin $t=0$):

$$\int_{0}^{+\infty}p(t)dt=R\int_{0}^{+\infty}I^{2}(t)dt=\frac{CV^{2}(t)}{2}$$

where $V(t=0)=V_{0}$ and $V(t=+\infty)=0$ (discharge).

Is the previous equation correct? One could solve the problem at any time an check that it is indeed valid... But one can also that the first time-derivative on both sides of the equation and get:

$$RI^{2}=\frac{1}{2}2CV(t)\frac{dV(t)}{dt}$$

That can be rewritten:

<img title="a title" alt="Alt text" src="../Figures/Cours 7/RC_check.png" width="250">

Where we can **identify** the two constitutive equations for the resistor ($RI(t)=V(t)$ in red) and for the capacitor ($I(t)=C\frac{dV(t)}{dt}$ in blue). This is simply a consistency check. A way to see that no energy is lost in our computations (what a relief).

Now, can we solve for $V(t)$? Exactly the same way as we did before, using charge conservation. It is left as an exercise to show that:

$$V(t)=V_{0}exp\left(\frac{-t}{\tau}\right)$$
