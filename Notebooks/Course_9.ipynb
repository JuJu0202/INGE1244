# The operational amplifier

After the battery, the resistor and the capacitor, the list of the fundamental building blocks of electronic compounds gets larger in this course thanks to the **operational amplifier** (Op Amp for short). As suggested by the name, this building block is used to amplify an electrical signal, as well as realizing operations on electrical signals. The signals considered are **voltage** signals. The Op Amp is so ubiquitous in electronics than it would be completely illusive to cover exhaustively its various usages. 

To start with, let us consider the Op Amp as a triangular-shaped **black box** with four inputs and one output. All the voltage signals presented on the following figure are defined with respect to the same potential reference (the ground). 

<img title="a title" alt="Alt text" src="../Figures/Cours 9/OpAmp_symbol.png" width="350">

Thanks to the Op Amp, the **difference** between $v^{+}$ and $v^{-}$ (on the left of the figure) gets **amplified** by a large factor, leading to the output voltage signal $v_{out}$. The equation governing *ideally* the Op Amp is given by:

$$v_{out}=G(v^{+}-v^{-})$$

where $G$ is called the **gain**.

- The "positive" "+" ($v^{+}$) input is called the "non-inverting" input. Indeed, a variation of $v^{+}$ at this input corresponds to a variation in the same "direction" at the output terminal $v_{out}$. By looking at the equation presented above, one can note that:

<div class="alert alert-block alert-info">
If $v^{+}$ increases then $v_{out}$ increases (with $v^{-}$ unchanged)
    
    
If $v^{+}$ decreases then $v_{out}$ decreases (with $v^{-}$ unchanged)
</div>

- The "negative" "-" ($v^{-}$) input is called the "inverting" input (in opposition with the positive input). Indeed:

<div class="alert alert-block alert-info">
If $v^{-}$ rises then $v_{out}$ decreases (with $v^{+}$ unchanged)
    
    
If $v^{-}$ decreases then $v_{out}$ increases (with $v^{+}$ unchanged)
</div>

- The "power inputs" provide power to the output signal. Ultimately, the output voltage is limited to the minimum and maximum value corresponding to the power supply voltages. The Op Amp can be symmetrically powered (e.g. $V+=+5V$ and $V-=-5V$) or with some asymmetry (e.g. $V+=+12V$ and $V-=-4V$).

Note that, in course 8, the Op Amp was presented as a **comparator** ("0" or "1" output depending on the input). It was used in a configuration where it always reached the minimum (the "0") or maximum (the "1") possible output value. Let us be a little more subtle with how the Op Amp can be used. 

## Static characteristics of the Op Amp

The gain in absense of all **feedback loop** (the concept of **feedback** is presented later in this notebook) is called the **open loop gain** (written $G_{0}$) and is defined as:

$$G_{0}=\frac{v_{out}}{v^{+}-v^{-}}$$

It is - by definition - a **dimensionless** quantity. One of the biggest subtlety of this course is that (for reasons that I hope will be clear to you at the end of the course):

<div class="alert alert-block alert-info">
The <b>ideal</b> open loop gain of an Op Amp should be <b>infinite</b>
</div>

Such clever requirement (that could appear weird to you at first glance) would date back to some of **Euler**'s considerations (eventhough, Karl Dale Swartzel Jr. is the inventor of the operational amplifier. He filed the patent for the 'summing amplifier' in 1941 when working at Bell Labs). 

Targeting the **infinity** is a huge challenge but nowadays Op Amps present typical values of $G_{0}$ between $10^{3}$ and $10^{6}$ (it can even reach $10^{8}$ in specific applications).  

<div class="alert alert-block alert-info">
Op Amp Golden rule #1: <b>no current</b> flows into the "+" nor the "-" *ideal* inputs (No matter the circuit). This rule is justified ex-cathedra.
</div>

Dealing with the many possible non-idealities of the *actual* Op Amp is the role of engineers and no such non-idealities are treated in this course.

### How resistors are implemented in silicon

Resistors are a key component of analog chips. Unfortunately, resistors in ICs are very inaccurate; the resistances can vary by 50% from chip to chip. Thus, analog ICs are designed so only the ratio of resistors matters, not the absolute values, since the ratios remain nearly constant from chip to chip (let us discuss that point).

The image below shows two resistors in the 741 op amp, formed using different techniques: 

<img title="a title" alt="Alt text" src="../Figures/Cours 9/resistors.png" width="400">

The resistor on the left (labeled R4) is formed from a **meandering strip** of P silicon (for the distinction between P doping and N doping - see later in the course), and is **about** 5KΩ. The resistor on the right is a **pinch resistor** and is about 50KΩ. In the pinch resistor, a layer of N silicon on top makes the conductive region much thinner (i.e. pinches it). This allows a much higher resistance for a given size. Both resistors are at the same scale below, but the pinch resistor has ten times the resistance. The tradeoff is the pinch resistor is much less accurate.

In the resistance formula, the meandering resistor "controls" the length of the resistor (can you tell who is the largest resistor between $R4$ and $R5$?) where the pinch resistor varies the resistivity ($\rho$) of the material itself by changing the doping.

### How capacitors are implemented in silicon

Capacitors are "easy" to implement on a die (could you propose a fabrication process? You can get inspired by course 3). Indeed, as the die is a conductor, all we need is a sandwich composed as follow: the die, an insulating layer (the dielectric) and a conductive layer atop (e.g. aluminum). The insultor is **silicon nitride**, a chemical compound of the elements silicon and nitrogen - $Si_{3}N_{4}$. It is the most thermodynamically stable and commercially important of the silicon nitrides. As a **passivation layer** for microchips, it is superior to silicon dioxide, as it is a significantly better diffusion barrier against water molecules and sodium ions, two major sources of **corrosion** and **instability** in microelectronics. It is also used as a dielectric between polysilicon layers in capacitors in analog chips.

<div class="alert alert-block alert-success">
<b>Business lies into the details</b>: A key innovation that led to the 741 was Fairchild's development of a new process for building capacitors on ICs using silicon nitride. Op amps before the 741 required an external capacitor to prevent oscillation, which was inconvenient. Dave Fullagar had the idea to put the compensation capacitor on the 741 chip using the new manufacturing process. Doing away with the external capacitor made the 741 extremely popular, either because engineers are lazy or because the reduced part count was beneficial.
</div>

On the 741, the only capacitor occupies the largest area of the chip (see the compound labeled C1 on the figure above). As we saw in previous course, the larger the area, the larger the capacitance. Also, silicon nitride presents a relative permittivity ($\epsilon$) of about 10 (bigger than the one of silicon dioxide - $\epsilon\sim 4$). The capacitance $C1$ is 25 $pF$ (what would be the value of $C1$ if silicon nitride was substituted by silicon dioxide?).

# How to use the Op Amp and for what?

Except in rare situations, the Op Amp is alsmost always used in a **closed loop** configuration to enable to exploit the concept of **feedback**. 

The keyword **feedback** is one of the trickiest in the course. It has many significations in many areas (Environnemental sciences, biology, education, business, engineering, physics, Math, etc.). The wikipedia definition will be used by default.

**Feedback** occurs when outputs of a system are routed back as inputs as part of a chain of cause-and-effect that forms a circuit or loop. The system can then be said to feed back into itself. The notion of cause-and-effect has to be handled carefully when applied to feedback systems:

<div class="alert alert-block alert-success">
Simple causal reasoning about a feedback system is difficult because the first system influences the second and second system influences the first, leading to a circular argument. This makes reasoning based upon cause and effect tricky, and it is necessary to <b>analyze the system as a whole</b>. As provided by Webster dictionnary, feedback in business is the transmission of evaluative or corrective information about an action, event, or process to the original or controlling source. (Karl Johan Åström and Richard M.Murray, Feedback Systems: An Introduction for Scientists and Engineers)
</div>

One can distinguish between two categories of feedback: 

- Positive feedback: If the signal feedback from output is in phase with the input signal, the feedback is called positive feedback.

- Negative feedback: If the signal feedback is of opposite polarity or out of phase by 180° (recall that $cos(\pi)=-1$ - this "minus" sign will be associated to the inverting pin "-" (multiplying a signal by $-1$)) with respect to input signal, the feedback is called negative feedback.

An example of positive feedback is discussed ex-cathedra: https://www.youtube.com/watch?v=B8aHLMsmsE8

The negative feedback is exploited in many circuit involving an Op Amp. Let us review the principal configurations.

### Config 1: The "follower"

The circuit is the following:

<div class="alert alert-block alert-info">
As stated above, it is necessary to analyze the system as a whole

As you can see, in this configuration, the output is **fed back** to the inverting pin of the Op Amp (both pins are shortcutted). 

This translate into the following "equation":

$$v^{-}=v_{out}$$

<div class="alert alert-block alert-danger">
The word equation needs here to be considered with great care. Indeed the value of $v^{-}$ is always extremely close to v_{out}. In that sense, the equation is *almost* right. **BUT** if at a time $t_{1}$, the output as a value of X (whatever value), this precise value $X$ will be reached, for $v^{-}$, at an instant $t_{2}$ with $t_{2}$ obviously greater than $t_{1}$. The **consequence** ($v^{-}$ reaching the value $X$) is always **after** and the **cause** ($v_{out}$ that reached the value $X$).
</div>

Recall that the other constitutive equation of the Op Amp is:

$$v_{out}=G(v^{+}-v^{-})$$

Suppose that we start an experiment with the circuit where, all voltages are 0 (ground) before time $t=0$. At $t=0$, a **step signal** is applied on the "+" pin, to reach the value $v^{+}$ (say +5V). The Op Amp "sees" a difference of 5V between the "+" pin and the "-" pin. As a causal reaction, the output voltage increases at a rate (called the **slew rate**) of typically 0.5 volts per micro seconds. 

<div class="alert alert-block alert-danger">
A frequent mistake appears often at this step of the reasoning. Students try to apply the constitutive equation $v_{out}=G(v^{+}-v^{-})$ and think that $v_{out}$ reaches instantaneously the value $v_{out}=G(v^{+}-0)$ (since $v^{-}$ is zero at the begining of the experiment). This is no physical reasoning.
</div>

Indeed, since the output is connected to the "-" pin. The Op Amp (at $t=0$) is asked to reach the "target voltage" (that is $Gv^{+}$) but, as $v_{out}$ increases, the voltage $v^{-}$ rises, *almost* instantaneously to the same value as $v_{out}$. As a consequence, the voltage difference between $v^{+}$ and $v^{-}$ is now **smaller**. It changes, as a consequence, the "target voltage". This continuous process reduces, to a vanishingly small value, the voltage difference between $v^{+}$ and $v^{-}$. The actual curve for $v_{out}$ depending on time depends on the Op Amp specific implementation but the following equations are always valid:

$$v_{out}=G(v^{+}-v^{-})$$

$$v^{-}=v_{out}$$

Solving for $v_{out}$ (do it as an exercise), one find:

$$v_{out}=\frac{G}{1+G}v^{+}$$

The beauty now comes from the fact that, since $G$ is huge, the 1 can be completely neglected and the numerator simplifies with the denominator to give:

$$v_{out}=v^{+}$$

Again, this equation is not exact, but is extremely precise for, e.g., $G\sim10^{5}$.

As you can see, the funny thing here is that the open loop gain is close to be infinite, but eventually, the closed loop gain is 1 since $$v_{out}=1\times v^{+}$$

What could it be used for? Since it simply copies the signal $v^{+}$ without any amplification. Discussed ex cathedra. 

The key thing to recall with this exercise is that:

<div class="alert alert-block alert-info">
Op Amp golden rule #2: In a negative feedback loop, the Op Amp reaches asymptotically a configuration where $v^{+}\sim v^{-}$. This fundamental equation can be used to solve easily the different circuit configurations that we will encounter.
</div>

By using the two golden rules of the Op Amp, we can illustrate how fast we can solve Op Amp circuits.
